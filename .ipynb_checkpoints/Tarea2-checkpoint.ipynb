```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
```

```python
# --- Aquí vamos a importar los datos y echarles una miradita
datos_raw_0=pd.read_csv('Cars93.csv');
datos_raw_0.head()
```

```python
# -- Aquí vamos a verificar que no hayan espacios vacíos en los datos
datos_raw_0.isnull().sum()
```

```python
label_d=np.array(datos_raw_0.columns[datos_raw_0.isnull().sum()!=0]);
print(label_d)
```

```python
# Hay 2 @#$&%@ columnas con huecos. Hay que rellenerlas
datos_raw1=datos_raw_0;
for j in range(len(label_d)):
    vec=np.array(datos_raw_0[label_d[j]]);
    a=np.array(datos_raw_0[label_d[j]].isnull());
    b=np.array(datos_raw_0[label_d[j]]);
    vec[a]=np.average(b[a==False]);
    datos_raw1[label_d[j]]=vec;
#fin for
datos_raw1.head()
```

```python
# -- Verificamos de nuevo que no hayan huecos
datos_raw1.isnull().sum()
```

```python
# creamos el vector de los nombres de las columnas por si se necesita
labels=np.array(datos_raw1.columns);
```

```python
from sklearn import preprocessing
```

```python
#print(datos_raw1['Manufacturer'])
a=datos_raw1['Manufacturer'];
a.dtype=='object'
labelStr=[];
for j in range(len(labels)):
    vecEvaluar=datos_raw1[labels[j]];
    if vecEvaluar.dtype=='object':
        labelStr.append(labels[j]);
    #fin if 
#fin for 
labelStr
```

```python
#Cambiamos los datos de 'string' a numéricos
pr1=preprocessing.LabelEncoder();
datosRaw2=datos_raw1;
for j in range(len(labelStr)):
    vectorReemplazar=pr1.fit_transform(datos_raw1[labelStr[j]]);
    datosRaw2[labelStr[j]]=vectorReemplazar;
#fin for 
datosRaw2.head()
```

```python
#lA Matriz de "datos" será la que se usará al final.
z=np.array(datosRaw2['Price'], dtype='int');
labelsN=[];
for i in range(len(labels)):
    if labels[i]!='Price':
        labelsN.append(labels[i]);
    #fin if
#fin for
print(labelsN)
datos=datosRaw2[labelsN];
datos.head()
```

```python
# Clasificación Visual
for k in range(len(labelsN)):
  fig=plt.figure()
  plt.plot(datos[labelsN[k]],z,'*')
  plt.xlabel(labelsN[k]);
  plt.ylabel('Precio');
  plt.show()
#fin for 
```

## Las variables que mas influyen en el precio son:
#### MPG city,MPG highway, DriveTrain, Cylinders, EngineSize, Horsepower, Rev.per.mile, Man.trans.avail, Fuel.tank.capacity, Length, Wheelbase, Weight,  

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=10)
X_train, X_test, Z_train, Z_test = train_test_split(np.array(datos),z)
model.fit(X_train,Z_train)
z_pred = model.predict(X_test)
```

```python
dife=np.zeros(len(Z_test))
dife[Z_test==z_pred]=1;
dife[Z_test!=z_pred]=0;
acc=sum(dife)/len(Z_test);
print(acc)
```

```python
importances = model.feature_importances_
ii = np.argsort(importances)[::-1]
np.array(labelsN)[ii]
```
